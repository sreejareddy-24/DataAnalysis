# FLIGHTS (WEATHER-TIME) DATASET

### 1. Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```


### 2. Load Dataset

```python
df = sns.load_dataset('flights')
df.head()
```


### 3. Explore Data

```python
print("Shape:", df.shape)
print(df.info())
print("\nMissing Values:\n", df.isnull().sum())
print("\nSummary:\n", df.describe())
```

## Insights:
### Dataset contains passenger counts by month and year.
### No missing values present.

### 4. Data Cleaning

```python
print("Duplicates:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

## Insights:
### Data is clean and ready for analysis.

### 5. Analysis with NumPy & Pandas

```python
avg_passengers = df['passengers'].mean()
print("Average Monthly Passengers:", avg_passengers)

growth_by_year = df.groupby('year')['passengers'].mean()
print("\nAverage Passengers by Year:\n", growth_by_year)
```

## Insights:
### Steady growth in passenger traffic from 1949 to 1960.

### 6. Visualizations

```python
sns.lineplot(x='year', y='passengers', data=df, marker='o')
plt.title("Average Passengers Over Years")
plt.show()

sns.lineplot(x='month', y='passengers', data=df)
plt.title("Seasonal Passenger Trends")
plt.show()

pivot = df.pivot(index="month", columns="year", values="passengers")
sns.heatmap(pivot, cmap='coolwarm')
plt.title("Monthly Passenger Heatmap")
plt.show()
```

## Insights:
### Passenger numbers peak during mid-year months (summer).
### Clear upward trend over the years.
