# CALIFORNIA HOUSING DATASET

### 1. Import Libraries


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing

sns.set_theme(style="whitegrid")
```

### 2. Load Dataset

```python
housing = fetch_california_housing(as_frame=True)
df = housing.frame
df.head()
```

### 3. Explore Data

```python
print("Shape:", df.shape)
print(df.info())
print("\nMissing Values:\n", df.isnull().sum())
print("\nSummary:\n", df.describe())

```

## Insights:
### Contains median income, house age, and other housing features.
### No missing values.



### 4. Data Cleaning

```python
print("Duplicates:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

## Insights:
### Data is clean and consistent.


### 5. Analysis with NumPy & Pandas


```python
mean_values = df.mean(numeric_only=True)
print("Average Values:\n", mean_values)

avg_income = df['MedInc'].mean()
print("\nAverage Median Income:", avg_income)
```

### 6. Visualizations


```python
sns.histplot(df['MedHouseVal'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Median House Value")
plt.show()

sns.scatterplot(x='MedInc', y='MedHouseVal', data=df)
plt.title("Median Income vs House Value")
plt.show()

sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
plt.title("Feature Correlations")
plt.show()
```

## Insights:
### Median income strongly correlates with house value.
### House value drops as distance from employment centers increases.
