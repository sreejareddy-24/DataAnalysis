# DIAMONDS DATASET

### 1. Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

### 2. Load Dataset

```python
df = sns.load_dataset('diamonds')
df.head()
```

### 3. Explore Data

```python
print("Shape:", df.shape)
print(df.info())
print("\nMissing Values:\n", df.isnull().sum())
print("\nSummary:\n", df.describe(include='all'))
```

## Insights:
### Dataset contains information about diamond characteristics and price.
### No missing values.



### 4. Data Cleaning


```python
print("Duplicates:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

## Insights:
### No duplicates found — data is clean.


### 5. Analysis with NumPy & Pandas


```python
mean_values = df.mean(numeric_only=True)
print("Average Values:\n", mean_values)

avg_price_cut = df.groupby('cut')['price'].mean()
print("\nAverage Price by Cut:\n", avg_price_cut)

avg_price_clarity = df.groupby('clarity')['price'].mean()
print("\nAverage Price by Clarity:\n", avg_price_clarity)
```

## Insights:
### Ideal cut diamonds don’t always mean higher price — color and clarity matter too.


### 6. Visualizations


```python
sns.histplot(df['price'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Diamond Prices")
plt.show()

sns.boxplot(x='cut', y='price', data=df, palette='Set2')
plt.title("Price by Cut Quality")
plt.show()

sns.scatterplot(x='carat', y='price', hue='cut', data=df)
plt.title("Carat vs Price by Cut")
plt.show()
```

## Insights:
### Price increases exponentially with carat weight.
### ‘Premium’ and ‘Ideal’ cuts dominate in higher price ranges.
