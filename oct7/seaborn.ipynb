# Seaborn - statistical data visualization
seaborn is a python library that builts on the top of matplotlib that makes the statistical data more easy and beautiful

## why we need it 
- for creating plots like box plot ,pair plot and violin plot
- automatically handles data from pandas dataframes

## where we use it 
- data analysis and reporting
- exploring relationships between variables
- machine learning for visualizing data sets and results 


```python
import seaborn as sns

df = sns.load_dataset("titanic")
df.head()
```

```python
## 4. Histograms & Countplots
# Histogram
import matplotlib.pyplot as plt
sns.histplot(df['age'].dropna(), bins=10, kde=True)
plt.title("Histogram of Age")
plt.show()
 
# Countplot
sns.countplot(x='class', data=df)
plt.title("Count of Passengers by Class")
plt.show()
```

```python
### 5. Scatter Plots
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
plt.title("Age vs Fare")
plt.show()
```

```python
### 6. Boxplot & Violin Plot
# Boxplot
sns.boxplot(x='class', y='age', data=df)
plt.title("Boxplot of Age by Class")
plt.show()
 
# Violin plot
sns.violinplot(x='class', y='age', hue='sex', data=df, split=True)
plt.title("Violin Plot of Age by Class and Sex")
plt.show()
 
```

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
```

```python
### 8. Heatmap
# Correlation heatmap (numeric columns only)
corr = iris.select_dtypes(include=['float64', 'int64']).corr()
 
import seaborn as sns
import matplotlib.pyplot as plt
 
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

```python
### 9. Customization & Themes
sns.set_theme(style="whitegrid")  # options: darkgrid, white, ticks, etc.
sns.histplot(df['age'].dropna(), bins=10, color='orange')
plt.title("Customized Histogram with Seaborn")
plt.show()
 
```

```python
sns.barplot(x='class', y='fare', data=df, hue='sex', ci='sd')
plt.title("Average Fare by Class and Sex")
plt.show()
 
 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# 1. Compare Passenger Survival by Class and Gender
plt.figure(figsize=(8,5))
sns.countplot(data=titanic, x='pclass', hue='survived')
plt.title("Passenger Survival by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(data=titanic, x='sex', hue='survived')
plt.title("Passenger Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# 2. Distribution of Passenger Ages
plt.figure(figsize=(8,5))
sns.histplot(titanic['age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# 3. Fare Distribution by Class
plt.figure(figsize=(8,5))
sns.boxplot(data=titanic, x='pclass', y='fare', palette='pastel')
plt.title("Fare Distribution by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()

# 4. Relationship Between Age and Fare
plt.figure(figsize=(8,5))
sns.scatterplot(data=titanic, x='age', y='fare', hue='pclass', palette='Set1')
plt.title("Relationship Between Age and Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()

# 5. Correlation Heatmap
corr = titanic[['survived','pclass','age','sibsp','parch','fare']].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Titanic Dataset")
plt.show()
```
